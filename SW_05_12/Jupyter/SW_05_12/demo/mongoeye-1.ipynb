# mongoeye

* https://github.com/mongoeye/mongoeye

```python
%%bash
type mongoeye
mongoeye --help
```

```python
%%bash
mongoeye --db mflix --col movies
```

```python
%%bash
mongoeye --db ships --col shipwrecks
```

```python
%%bash
mongoeye --db ships --col shipwrecks --sample random:1000
```

```python
%%bash
mongoeye --db ships --col shipwrecks --sample all 
```

```python
%%bash
mongoeye --db ships --col shipwrecks --sample all --most-freq 5 --count-unique --format json
```

```python
%%bash
mongoeye --db ships --col shipwrecks --sample all --most-freq 5 --count-unique --format json | 
  mongoimport --db admin_stats --collection ships
```
